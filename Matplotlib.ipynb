### Basic Graph & Trying different features for graphs



```python
x = [1, 2, 4]
y = [5,7,8]
```

```python
plt.plot(x,y, label = 'xy',color = 'g', linewidth = 5, marker = '>', markersize = 10, markeredgecolor = 'red')
plt.title('Hello Graph :)', fontdict={'fontname': 'Comic Sans MS', 'fontsize':26})
plt.xlabel('x')
plt.ylabel('y')
plt.xticks([1, 2, 3, 4])
plt.legend()
plt.show()
```
